```python
from datadownloader.Economy import Currency

import matplotlib.pyplot as plt
```

# Get exchange rate between currencies (Default base is USD)

```python
usd_cad = Currency('CAD')
print(usd_cad.exchange_symbol)
usd_cad_hist = usd_cad.get_history()
print(usd_cad_hist)
plt.plot(usd_cad_hist)
plt.show()
```

```python
eur_cad = Currency('CAD', base_currency='EUR')
print(eur_cad.exchange_symbol)
eur_cad_hist = eur_cad.get_history()
print(eur_cad_hist)
plt.plot(eur_cad_hist)
plt.show()
```
